```python
# Retrosheet doesn't declare a winner so define new columns for winner, and for vis and home winlose

gls['winner'] = ''
gls.loc[gls[(gls.home_score > gls.vis_score)].index,'winner'] = gls.home
gls.loc[gls[(gls.home_score < gls.vis_score)].index,'winner'] = gls.vis

# Ties = 0
# Wins = 1
# Losses = -1

gls['vis_winlose'] = 0
gls.loc[gls[(gls.home_score > gls.vis_score)].index,'vis_winlose'] = -1
gls.loc[gls[(gls.home_score < gls.vis_score)].index,'vis_winlose'] = 1

gls['home_winlose'] = 0
gls.loc[gls[(gls.home_score > gls.vis_score)].index,'home_winlose'] = 1
gls.loc[gls[(gls.home_score < gls.vis_score)].index,'home_winlose'] = -1
```

## Question 1: What are the most common W/L sequences for game sequences of length n?

```python
def wlstring(a):
    lu = {1 : 'W', 0 : 'T', -1 : 'L'}
    return ''.join([lu[x] for x in a])
```

```python
for w in range(2,20):
    
    # for storing W/L seqs
    
    wl_seqs = []

    # group data by season and team
    
    for i,r in games[games.season >= 1903].groupby(['season','team']):
        
        # get raw winlose vector
        wl_season = wlstring(r.winlose.values)
        
        # loop through starting positions
        for j in range(0,len(wl_season)- w + 1):
            
            # get subseq
            wl_seq = wl_season[j:j+w]
            
            # ignore streaks with ties
            
            if 'T' not in wl_seq:
                
                # keep list of wl_seqs
                wl_seqs.append(wl_seq)

    # count occurences of wl_seqs
    c_wl_seqs = Counter(wl_seqs)

    # list two most common and two least common wl_seqs for each W
    for k in c_wl_seqs.most_common(2):
        print (w,k[0],k[1])
    for k in c_wl_seqs.most_common()[-2:]:
        print (w,k[0],k[1])    
    print ("")
```

### Answer 1: Long losing streaks are the most common sequence of wins and losses and long winning streaks the second most common for essentially all streak lengths until the numbers become too small to be relevant. 

This isn't really surprising given that our expectation is that few teams are expected to win precisely half of their games and if p(win) > .5 a long win streak is most likely sequence of n individual events and if p(win) < .5 a long losing streak is the most likely sequence of n events. 

Note do not confuse this with the question of how many wins one expects during a streak of length n. This is about the precise sequence of wins and losses.

## Question 2: How much more common are winning or losing streaks than alternating W/L streaks for each length n?

```python
# code is same -- output different

# store for plotting 

x = []
y = []


for w in range(2,20):
    
    # for storing W/L seqs
    
    wl_seqs = []

    # group data by season and team
    
    for i,r in games[games.season >= 1903].groupby(['season','team']):
        
        # get raw winlose vector
        wl_season = wlstring(r.winlose.values)
        
        # loop through starting positions
        for j in range(0,len(wl_season)- w + 1):
            
            # get subseq
            wl_seq = wl_season[j:j+w]
            
            # ignore streaks with ties
            
            if 'T' not in wl_seq:
                
                # keep list of wl_seqs
                wl_seqs.append(wl_seq)

    # count occurences of wl_seqs
    c_wl_seqs = Counter(wl_seqs)
    
    w_str = ''.join(['W'] * w)
    l_str = ''.join(['L'] * w)
    
    alt1_str = ''.join(['WL'] * int(w/2) + ['W'] * (w % 2))
    alt2_str = ''.join(['LW'] * int(w/2) + ['L'] * (w % 2))
    
    n_homogenous = c_wl_seqs[w_str] + c_wl_seqs[l_str]
    n_alternating = c_wl_seqs[alt1_str] + c_wl_seqs[alt2_str]
    
    if (n_alternating > 0):
        print ("%3d %6.1f" % (w,n_homogenous/n_alternating))
        
        x.append(w)
        y.append(n_homogenous/n_alternating)
```

### Answer 2: For streaks over 10 games (where one might start to notice) you start at 3x more likely to observe a homogenous streak compared to an alternating one, and that keeps going up.


```python
plt.plot(x,y)
plt.xlabel("Length of Streak")
plt.ylabel("p(homogenous)/p(alternating)")
```

## Question 3: What is the shortest unique W/L sequence in baseball history?

```python
# First lets figure out how high n has to be to get unique streaks

for w in range(2,20):
    
    # for storing W/L seqs
    
    wl_seqs = []

    # group data by season and team
    
    for i,r in games.groupby(['season','team']):
        
        # get raw winlose vector
        wl_season = wlstring(r.winlose.values)
        
        # loop through starting positions
        for j in range(0,len(wl_season)- w + 1):
            
            # get subseq
            wl_seq = wl_season[j:j+w]
            
            # ignore streaks with ties
            
            if 'T' not in wl_seq:
                
                # keep list of wl_seqs
                wl_seqs.append(wl_seq)

    # count occurences of wl_seqs
    c_wl_seqs = Counter(wl_seqs)
    
    # output w and the counts of the least common wl_seq
    print (w,c_wl_seqs.most_common()[-1][1])
```

So you don't find a unique sequence until you get to streaks of length 15. What are those streaks?

```python
# First lets figure out how high n has to be to get unique streaks

w = 15
    
# for storing W/L seqs

wl_seqs = []

# group data by season and team

for i,r in games.groupby(['season','team']):

    # get raw winlose vector
    wl_season = wlstring(r.winlose.values)

    # loop through starting positions
    for j in range(0,len(wl_season)- w + 1):

        # get subseq
        wl_seq = wl_season[j:j+w]

        # ignore streaks with ties

        if 'T' not in wl_seq:

            # keep list of wl_seqs
            wl_seqs.append(wl_seq)

# count occurences of wl_seqs
c_wl_seqs = Counter(wl_seqs)

# go through again and extract info from streaks with count = 1

data = []

for i,r in games.groupby(['season','team']):

    # get raw winlose vector
    wl_season = wlstring(r.winlose.values)

    # loop through starting positions
    for j in range(0,len(wl_season)- w + 1):

        # get subseq
        wl_seq = wl_season[j:j+w]

        # ignore streaks with ties

        if 'T' not in wl_seq:

            if c_wl_seqs[wl_seq] == 1:
                
                # get info
                rr = games[(games.season == i[0]) & (games.team == i[1]) & (games.gamenum == (j+1))].iloc[0]
                dt = str(rr.date)
                
                data.append([i[0],i[1],dt[4:6],dt[6:],dt[0:4],wl_seq])

wl_short_unique = pd.DataFrame(data=data,columns=['season','team','month','day','year','wl_seq'])
```

## Question 4: When do you start to see streaks that have never been observed? 

```python
# Are there any sequences of 14 or fewer games that have never occurred?

for w in range(2,20):
    
    # for storing W/L seqs
    
    wl_seqs = []

    # group data by season and team
    
    for i,r in games.groupby(['season','team']):
        
        # get raw winlose vector
        wl_season = wlstring(r.winlose.values)
        
        # loop through starting positions
        for j in range(0,len(wl_season)- w + 1):
            
            # get subseq
            wl_seq = wl_season[j:j+w]
            
            # ignore streaks with ties
            
            if 'T' not in wl_seq:
                
                # keep list of wl_seqs
                wl_seqs.append(wl_seq)

    # count occurences of wl_seqs
    c_wl_seqs = Counter(wl_seqs)
    
    # expected number of streaks = 2 ** w
    
    # output w and the counts of the least common wl_seq
    print (w,len(c_wl_seqs), 2 ** w)
```

### Answer 4: Until you get to length 15, all possible sequences have been observed. 

### For length 15 there is one that has not been observed. What is it?

```python
# Not sure there's a python tool to generate all possible samples so let's just do this crudely

# First lets find all the ones that have been observed

w = 15
    
# for storing W/L seqs

wl_seqs = []

# group data by season and team

for i,r in games.groupby(['season','team']):

    # get raw winlose vector
    wl_season = wlstring(r.winlose.values)

    # loop through starting positions
    for j in range(0,len(wl_season)- w + 1):

        # get subseq
        wl_seq = wl_season[j:j+w]

        # ignore streaks with ties

        if 'T' not in wl_seq:

            # keep list of wl_seqs
            wl_seqs.append(wl_seq)

# count occurences of wl_seqs
c_wl_seqs = Counter(wl_seqs)

# now go through and use the permutation tool to generate all possible permutations with different numbers of wins

for wins in range(0,w+1):
    for c in combinations(range(0,w),wins):
        wl_seq = ['L'] * w
        for s in c:
            wl_seq[s] = 'W'
        if ''.join(wl_seq) not in c_wl_seqs:
            print (''.join(wl_seq))
```

### Answer 4: The only streak of length 15 never to occur is WWLLLWWLWLLLWWL

## Question 5: What is the longest sequence to have happened more than once?

```python

# We'll start this at 20
for w in range(20,40):
    
    # for storing W/L seqs
    
    wl_seqs = []

    # group data by season and team
    
    for i,r in games.groupby(['season','team']):
        
        # get raw winlose vector
        wl_season = wlstring(r.winlose.values)
        
        # loop through starting positions
        for j in range(0,len(wl_season)- w + 1):
            
            # get subseq
            wl_seq = wl_season[j:j+w]
            
            # ignore streaks with ties
            
            if 'T' not in wl_seq:
                
                # keep list of wl_seqs
                wl_seqs.append(wl_seq)

    # count occurences of wl_seqs
    c_wl_seqs = Counter(wl_seqs)
    
    # output w and the counts of the most common wl_seq
    # we're looking for when this drops to 1 
    
    print (w,c_wl_seqs.most_common()[0][1])
```

```python
# go through again and extract info from streaks with count = 2

w = 36

wl_seqs = []

# group data by season and team

for i,r in games.groupby(['season','team']):

    # get raw winlose vector
    wl_season = wlstring(r.winlose.values)

    # loop through starting positions
    for j in range(0,len(wl_season)- w + 1):

        # get subseq
        wl_seq = wl_season[j:j+w]

        # ignore streaks with ties

        if 'T' not in wl_seq:

            # keep list of wl_seqs
            wl_seqs.append(wl_seq)

# count occurences of wl_seqs
c_wl_seqs = Counter(wl_seqs)

data = []

for i,r in games.groupby(['season','team']):

    # get raw winlose vector
    wl_season = wlstring(r.winlose.values)

    # loop through starting positions
    for j in range(0,len(wl_season)- w + 1):

        # get subseq
        wl_seq = wl_season[j:j+w]

        # ignore streaks with ties

        if 'T' not in wl_seq:

            if c_wl_seqs[wl_seq] == 2:
                
                # get info
                rr = games[(games.season == i[0]) & (games.team == i[1]) & (games.gamenum == (j+1))].iloc[0]
                dt = str(rr.date)
                
                data.append([i[0],i[1],dt[4:6],dt[6:],dt[0:4],wl_seq])

wl_long_shared = pd.DataFrame(data=data,columns=['season','team','month','day','year','wl_seq'])

wl_long_shared
```
